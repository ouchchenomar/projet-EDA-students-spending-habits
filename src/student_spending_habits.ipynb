## introduction

* Cette dataset contient des données fictives représentant les habitudes de dépenses de 1000 étudiants issus de divers groupes démographiques et de milieux académiques.

## Visualisation des Données :

```python
plt.figure(figsize=(12,6))
sns.barplot(x='major', y='monthly_income', data=dt, order=major_order)
plt.xlabel('student major')
plt.ylabel('monthaly income')


plt.title('income of student')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='major', y='financial_aid', data=dt, order=major_order)
plt.xlabel('student major')
plt.ylabel('financial_aid')


plt.title('income of student from finacial aid')
plt.show()
```

```python
# Aggregate expenses by category
expense_categories = ['tuition', 'housing',
            'food', 'transportation', 'books_supplies', 'entertainment',
            'personal_care', 'technology', 'health_wellness', 'miscellaneous']
total_expenses = data[expense_categories].sum()
```

```python
# Visualize expense distribution
plt.figure(figsize=(12, 10))
total_expenses.plot(kind='pie', autopct='%1.1f%%', startangle=150)
plt.title('Expense Distribution Across Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

d'apres la figure on conclue que
*  Les frais de scolarité représentent la plus grande part des dépenses, ce qui est attendu étant donné le coût élevé de l'éducation.
*  Les dépenses dans ces catégories Technologie  reflètent l'engagement des étudiants dans leurs études et leur utilisation de la technologie.


```python
dt['Total Expenses'] = dt[ ['tuition', 'housing',
            'food', 'transportation', 'books_supplies', 'entertainment',
            'personal_care', 'technology', 'health_wellness', 'miscellaneous']].sum(axis=1)
```

```python
# Calculate total monthly income and total financial aid for each student
dt['Total Monthly Income'] = dt['monthly_income'] + dt['financial_aid']
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(dt['Total Monthly Income'],dt['Total Expenses'], alpha=0.5)
plt.title('Total Monthly Income + Financial Aid vs Total Expenses')
plt.ylabel('Total Monthly Income + Financial Aid ($)')
plt.xlabel('Total Expenses ($)')
plt.grid(True)
plt.show()
```

## Conclusions sur les Catégories de Dépenses des Étudiants :

### Catégorie à Revenu Moyen et Dépenses Modérées :
* La majorité des étudiants semblent avoir un revenu moyen et dépenser de manière modérée. Cela suggère une gestion équilibrée des finances, où les dépenses sont adaptées aux revenus disponibles.
### Catégorie à Revenu Élevé et Faibles Dépenses :
* Certains étudiants ayant des revenus élevés ont tendance à dépenser moins. Cela peut indiquer une gestion financière efficace ou des priorités de dépenses différentes.
### Catégorie à Faible Revenu et Fortes Dépenses :
* D'autres étudiants à faible revenu peuvent avoir des dépenses plus élevées. Cela pourrait être dû à des besoins essentiels ou à des charges financières telles que les frais de scolarité ou le logement.
### Catégorie à Faible Revenu et Faibles Dépenses :
* Certains étudiants ayant un faible revenu ont tendance à dépenser peu d'argent. Cela peut refléter une gestion prudente des finances en raison de ressources limitées.
### Catégorie à Revenu Élevé et Fortes Dépenses :
* Enfin, il existe une catégorie d'étudiants avec des revenus élevés qui dépensent beaucoup. Cela peut être attribuable à des choix de vie ou à des possibilités financières plus importantes.

## Corrélation:

```python
# Calculate the correlation coefficients
correlation = dt[['tuition', 'housing',
            'food', 'transportation', 'books_supplies', 'entertainment',
            'personal_care', 'technology', 'health_wellness', 'miscellaneous', 'Total Expenses']].corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

### Interprétations Clés de correlation :
* Corrélation Positive Forte : Les cellules en rouge foncé indiquent des corrélations positives fortes. Cela signifie que lorsque une variable augmente, l'autre variable a tendance à augmenter également.

* Corrélation Négative Forte : Les cellules en bleu foncé indiquent des corrélations négatives fortes. Cela signifie que lorsque une variable augmente, l'autre variable a tendance à diminuer.

* Absence de Corrélation : Les cellules blanches (proches de 0) indiquent peu ou pas de relation linéaire entre les variables.
